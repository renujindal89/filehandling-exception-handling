```python
When to use what?
✔ Use write() when:

Writing one line or one string at a time

Building file content manually

✔ Use writelines() when:

You already have a list of strings

You want to write many lines at once
```

```python
#create a csv file using list
list--->csv
csv--->list
```

```python
import csv

# Data to write (list of lists or list of tuples)
data = [
    ['name', 'age', 'city'],  # header row
    ['Alice', 30, 'New York'],
    ['Bob', 25, 'Los Angeles'],
    ['Charlie', 35, 'Chicago']
]

# Open a new file in write mode
with open('people5.csv', 'w') as f:
    writer = csv.writer(f)

    # Write all rows
    writer.writerows(data)
```

```python
# read a csv file  
```

```python
#rows as lists)
import csv

with open('people5.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)  # each row is a list of strings
# shows empty lists [] after every row.
#This usually happens because the CSV file has extra blank lines between the rows.        
```

```python
# read csv in pandas
import pandas as pd
df=pd.read_csv("people5.csv")
df
```

```python
#rows as dictionaries
import csv

with open('people5.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row)  # each row is a dict with column names as keys
```

```python
Using JSON (BEST method for lists/dicts)

If you are storing Python lists or dictionaries, JSON is the safest and easiest way.
```

```python
import json

data = ["apple", "banana", "cherry"]

with open("data.json", "w") as f:
    json.dump(data, f) 
```

```python
#Write list or dict to a file
import json

data = {
    "name": "Alice",
    "age": 21,
    "marks": [88, 92, 79]
}

with open("data1.json", "w") as f:
    json.dump(data, f)  # {"name": "Alice", "age": 21, "marks": [88, 92, 79]}
```

```python
#Write list or dict to a file
import json

data = {
    "name": "Alice",
    "age": 21,
    "marks": [88, 92, 79]
}

with open("data3.json", "w") as f:
    json.dump(data, f, indent=4)
```

```python
#Read list/dict from a file
import json

with open("data3.json", "r") as f:
    data = json.load(f)

print(data)
```
