```python
try:
    x = int("abc")
except ValueError:
    print("Conversion failed!")
```

```python
#ZeroDivisionError
# When dividing by zero

try:
    x = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
    
```

```python
#ValueError
#Wrong value type

try:
    num = int("abc")
except ValueError:
    print("Invalid input")
    
```

```python
#TypeError
#Wrong data type operation

try:
    result = "10" + 5
except TypeError:
    print("Type mismatch")
    
```

```python
#IndexError
#Accessing invalid index

try:
    arr = [1, 2, 3]
    print(arr[5])
except IndexError:
    print("Index out of range")
    
```

```python
#KeyError
#Accessing missing key in dictionary

try:
    d = {"name": "Renu"}
    print(d["age"])
except KeyError:
    print("Key not found")
    
```

```python
#FileNotFoundError
#File does not exist

try:
    f = open("data.txt")
except FileNotFoundError:
    print("File not found")
    
```

```python
#AttributeError
#Attribute doesn’t exist

try:
    x = 10
    x.append(5)
except AttributeError:
    print("Invalid attribute")
    
```

```python
# NameError
#Variable not defined

try:
    print(a)
except NameError:
    print("Variable not defined")
    
```

```python
#Generic Exception (Catch All)
try:
    x = int("abc")
except Exception as e:
    print("Error:", e)
```

```python
#Handles any type of error
#Multiple Exceptions
try:
   
    y = 10 / 2
except ValueError:
    print("Value error")
except ZeroDivisionError:
    print("Divide by zero")
    
```

```python
#Finally Block
#Always runs (cleanup code)

try:
    f = open("file.txt")
except:
    print("Error")
finally:
    print("Done")
```

```python
try: Code that might cause an error

except: Handles the error

else: Runs if no error occurs

finally: Runs always (cleanup code)
```

```python
try:
    x = 10 / 2
except ZeroDivisionError:
    print("Cannot divide by zero!")
else:
    print(x)
    print("No error occurred.")
finally:
    print("This runs no matter what.")
```

```python
try:
    # taking input from user
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))

    # division
    result = num1 / num2

    print("Result:", result)

# handling division by zero
except ZeroDivisionError:
    print("Error: Cannot divide by zero!")

# handling invalid input (optional but useful)
except ValueError:
    print("Error: Please enter valid numbers!")
except TypeError:
    print("Error: you have entered  character!")    

# always runs
finally:
    result2 = num1 + num2
    print(result2)
    print("Program executed successfully.")
```
