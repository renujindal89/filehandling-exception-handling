# adding a method

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    # This is a method
    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)

s1 = Student("Renu", 36)
s1.display()   # calling method
```

# Encapsulation

```python
It means wrapping data (variables) and methods (functions) together into a single unit (class)
```

```python
#Public Members
class Person:
    def __init__(self, name, age):
        self.name = name      # public
        self.age = age        # public

p = Person("Renu", 25)
print(p.name)
print(p.age)


#✔ Accessible from outside the class
```

```python
#private Accessible only inside the class
class Person:
    def __init__(self, name):
        self.__name = name   # private  Cannot access directly outside class

    def show_name(self):
        print(self.__name)
    def new(self):
        print("hello")
```

```python
p1 = Person("Renu")

print(p1.__name)   # ❌ Error
```

```python
p = Person("Renu")
p.show_name()        # works
```

# inheritance

```python
# inheritance 
when one class derived the property and method of other class 
parent -----> child (good values inherit )
class car(): # parent class
    start method
    stop method
    color attribute
class toyotacar(car): # child class,derived class
    
```

```python
# single level -> parent->child
class car():
    def start(self):
        print(" car started")
    def stop(self):
        print(" car stopped")  
class tyotacar():
    def __init__(self,name):
        self.name=name
car1=tyotacar("fortuner")  
car1.start()
car1.stop()
car1.name       
```

```python
# single level -> parent->child
class car():
    def start(self):
        print(" car started")
    def stop(self):
        print(" car stopped")  
class tyotacar(car):
    def __init__(self,name):
        self.name=name
car1=tyotacar("fortuner")  
car1.start()
car1.stop()
car1.name                       # sum()
```

```python
# multiple inheritance  #parent-------->parent----------> child
class a():
    var1="welcome to class a"
class b():
    var2="welcome to class b"
class c(a,b):
    var3="welcome to class c"
c1=c()
print(c1.var3)
print(c1.var2)
print(c1.var1)
```

# polymorphism

```python
The same function/method name behaves differently depending on the object or context.
```

```python
# polymorphism
print(1+2) #3
print("renu"+"jindal") # concatination
print([1,2,3]+[4,5,6]) # merge
```

```python
operator overloading
when the same operator is allowed to have  different meaning according to the context
a+b # a.__add__(b)
a-b    a.__sub__(b)
a*b
a/b
```

```python
class Number:
    def __init__(self, value):
        self.value = value

    # Overloading the + operator
    def __add__(self, other):
        return self.value + other.value

a = Number("renu")
b = Number("jindal")
print(a + b)  # 15
```

```python
a = Number(10)
b = Number(5)
print(a + b)  # 15
```

```python
a = Number([1,2,3])
b = Number([5,6,7])
print(a + b)  # 15
```
